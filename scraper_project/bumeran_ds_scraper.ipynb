```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time


# Configurar el navegador (usar Chrome en este caso)
driver = webdriver.Chrome()

# Ir a la página web donde están los botones
driver.get(url)  # Reemplaza con la URL real

# Esperar a que la página cargue completamente
time.sleep(3)  

# Aplicar los filtros (según tu código)
def click_button(text):
    try:
        button = driver.find_element(By.XPATH, f"//button[contains(text(), '{text}')]")
        button.click()
        time.sleep(2)
    except Exception as e:
        print(f"No se encontró el botón: {text} - {e}")

click_button("Fecha de publicación")
click_button("Menor a 15 días")
click_button("Área")
click_button("Tecnología, Sistemas y Telecomunicaciones")
click_button("Subárea")
click_button("Programación")
click_button("Departamento")
click_button("Lima")
click_button("Carga horaria")
click_button("Full-time")

# Obtener el HTML de la página
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

# Extraer todos los links dentro de <a> con href
links = []
for a_tag in soup.find_all("a", href=True):
    link = a_tag["href"]
    # Filtrar solo los enlaces que contienen ofertas de empleo
    if "/empleos/" in link:
        links.append("https://www.bumeran.com.pe" + link)  # Agregar el dominio

# Cerrar el WebDriver
driver.quit()

# Imprimir los enlaces extraídos
for link in links:
    print(link)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

# Configurar Selenium
driver = webdriver.Chrome()
driver.maximize_window()

# Lista de URLs de ofertas de trabajo
urls = links  # Asegúrate de definir 'links' antes de usarlo

# Función para extraer el distrito correctamente
def get_distrito(driver):
    xpaths = [
        "/html/body/div[1]/div/div[2]/div[2]/div/div[2]/div[1]/div[2]/div[1]/div[1]/div[2]/div/div/div[1]/div[2]/div/div/li/a/h2",
        "/html/body/div[1]/div/div[2]/div[2]/div/div[2]/div[1]/div[2]/div[1]/div[1]/div[2]/div/div[1]/div[1]/div[2]/div/div/li/a/h2",
    ]
    for xpath in xpaths:
        try:
            distrito_texto = WebDriverWait(driver, 5).until(
                EC.presence_of_element_located((By.XPATH, xpath))
            ).text

            # Limpiar el texto para eliminar "Actualizado hace X horas" y "Nuevo"
            distrito_limpio = distrito_texto.split(",")[0].strip()
            return distrito_limpio
        except:
            continue
    return "Distrito no encontrado"

# Función para extraer el modo de trabajo
def get_modo_trabajo(driver):
    xpaths = [
        "/html/body/div[1]/div/div[2]/div[2]/div/div[2]/div[1]/div[2]/div[1]/div[1]/div[2]/div/div[1]/div[4]/div/ul/div[1]/li[1]/a/p",
        "/html/body/div[1]/div/div[2]/div[2]/div/div[2]/div[1]/div[2]/div[1]/div[1]/div[2]/div/div/div[4]/div/ul/div[1]/li[1]/a/p",
        "/html/body/div[1]/div/div[2]/div[2]/div/div[2]/div[1]/div[2]/div[1]/div[1]/div[2]/div/div/div[5]/div/ul/div[1]/li[1]/a/p",
        "/html/body/div[1]/div/div[2]/div[2]/div/div[2]/div[1]/div[2]/div[1]/div[1]/div[2]/div/div[1]/div[5]/div/ul/div[1]/li[1]/a/p",
    ]
    for xpath in xpaths:
        try:
            modo = WebDriverWait(driver, 5).until(
                EC.presence_of_element_located((By.XPATH, xpath))
            ).text
            return modo
        except:
            continue
    return "Modo de trabajo no encontrado"

# Lista para almacenar datos
job_data = []

for url in urls:
    try:
        driver.get(url)
        time.sleep(3)  # Esperar que la página cargue completamente

        # Extraer título del trabajo
        job_title = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.TAG_NAME, "h1"))
        ).text

        # Extraer descripción hasta "Beneficios"
        description_element = driver.find_element(By.XPATH, "//*[@id='ficha-detalle']")
        full_description = description_element.text.split("\n")

        # Filtrar hasta "Beneficios" (si existe)
        description_text = []
        for line in full_description:
            if "Beneficios" in line:
                break
            description_text.append(line)

        description = "\n".join(description_text)

        # Extraer distrito y modalidad de trabajo con las funciones optimizadas
        district = get_distrito(driver)
        work_mode = get_modo_trabajo(driver)

        # Guardar en la lista
        job_data.append([job_title, description, district, work_mode])

        print(f"Datos extraídos de: {url}")

    except Exception as e:
        print(f"Error en {url}: {e}")

# Cerrar el navegador
driver.quit()

# Crear un DataFrame y exportar a Excel
df_jobs = pd.DataFrame(job_data, columns=["Job Title", "Description", "District", "Work Mode"])
df_jobs.to_excel("ds_job_listings_bumeran.xlsx", index=False, engine="openpyxl")

print("Todos los datos han sido guardados en ds_job_listings_bumeran.xlsx")
```
